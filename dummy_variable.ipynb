```python
import pandas as pd

df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
```

```python
concated_df = pd.concat([df,dummies],axis='columns')
concated_df
```

```python
final = concated_df.drop(['town','west windsor'],axis='columns')
final
```

```python
x = final.drop('price' , axis = 'columns')
x
```

```python
y = final.price
y
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[2600,0,1]])
```
