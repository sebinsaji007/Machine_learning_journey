```python
import numpy as np

# input data
x = np.array([1, 2, 3, 4, 5])
# output data
y = np.array([5, 7, 9, 11, 13])

# initialize parameters
m = 0
b = 0

# learning rate
alpha = 0.001

# number of iterations
num_iterations = 1000

# gradient descent
for i in range(num_iterations):
    # calculate predictions
    y_pred = m * x + b

    # calculate error
    error = y - y_pred

    # calculate cost
    cost = np.sum(error ** 2) / len(x)

    # print cost
    print(f"Iteration {i}: Cost={cost}")

    # calculate gradients
    m_grad = -2 * np.sum(error * x) / len(x)
    b_grad = -2 * np.sum(error) / len(x)

    # update parameters
    m = m - alpha * m_grad
    b = b - alpha * b_grad

print(f"Optimal values: m={m}, b={b}")
```
