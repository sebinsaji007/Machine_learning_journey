```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience=df.experience.fillna('zero')
df.experience
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
```

```python
new_column_names = {
    "test_score(out of 10)": "test_scores",
    "interview_score(out of 10)":"interview_score",
    "salary($)":"salary"
}
df.rename(columns=new_column_names, inplace=True)
```

```python
df
```

```python
med_test_scores=df.test_scores.median()
df.test_scores=df.test_scores.fillna(med_test_scores)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['experience','test_scores','interview_score']],df.salary)
```

```python
reg.predict([[7,8,9]])
```
